```python
import numpy as np
import pandas as pd 
from scipy import signal
import matplotlib.pyplot as plt
import math
import os
import glob
import seaborn as sns; sns.set()
```

```python
raw_dir = "/content/drive/My Drive/ML Data/Offline Mode Classification Result/"
data_path = raw_dir + "gaitphase.xlsx"
result1 = pd.read_excel(data_path, header=0)
```

```python
raw_dir = "/content/drive/My Drive/ML Data/Offline Mode Classification Result/"
data_path = raw_dir + "speed.xlsx"
result2 = pd.read_excel(data_path, header=0)
```

```python
palette = [(0.0, 0.10980392156862745, 0.4980392156862745), (0.5490196078431373, 0.03137254901960784, 0.0)]
image1 = sns.lineplot(x="Time (min)", y ="Gait Phase Estimation Error (%)", hue = "Model", ci="sd" ,data = result1, linewidth = 5, palette= palette)
image1.set(ylim=(0, 10))
image1.tick_params(labelsize=16)
image1.set_xlabel("Time (min)", fontsize=20)
image1.set_ylabel("Gait Phase\nEstimation Error (%)", fontsize=20)
image1.legend(labels=['Static', 'Adapted'], fontsize=18)
sns.despine()
```

```python
palette = [(0.0, 0.10980392156862745, 0.4980392156862745), (0.5490196078431373, 0.03137254901960784, 0.0)]
image2 = sns.lineplot(x="Time (min)", y ="Speed Estimation Error (%)", hue = "Model", ci="sd" ,data = result2,  linewidth = 5, palette= palette)
image2.set(ylim=(0, 0.2))
image2.tick_params(labelsize=16)
image2.set_xlabel("Time (min)", fontsize=20)
image2.set_ylabel("Speed\nEstimation Error (m/s)", fontsize=20)
image2.legend(labels=['Static', 'Adapted'], fontsize=18)
sns.despine()
```

```python
image1.figure.savefig(raw_dir+'gaitphase.svg', format='svg')
image2.figure.savefig(raw_dir+'speed.svg', format='svg')
```
