```python
import numpy as np
import pandas as pd 
from scipy import signal
import matplotlib.pyplot as plt
import math
import os
import glob
import seaborn as sns; sns.set()
```

```python
raw_dir = "/content/drive/My Drive/ML Data/Offline Mode Classification Result/"
data_path = raw_dir + "gaitphase.xlsx"
result1 = pd.read_excel(data_path, header=0)
```

```python
raw_dir = "/content/drive/My Drive/ML Data/Offline Mode Classification Result/"
data_path = raw_dir + "speed.xlsx"
result2 = pd.read_excel(data_path, header=0)
```

```python
image1 = sns.lineplot(x="Time (min)", y ="Gait Phase Estimation Error (%)", hue = "Model", ci="sd" ,data = result1)
sns.set(style="white", context="talk", palette="deep")
sns.set(style="ticks")
image1.set_xticks(np.arange(0, 10, 2))
sns.despine()
```

```python
image2 = sns.lineplot(x="Time (min)", y ="Speed Estimation Error (%)", hue = "Model", ci="sd" ,data = result2)
sns.set(style="white", context="talk", palette="deep")
sns.set(style="ticks")
image2.set_yticks(np.arange(0.05, 0.25,0.05))
image2.set_xticks(np.arange(0, 10, 2))
sns.despine()
```

```python
image1.figure.savefig(raw_dir+'gaitphase.png', format='png', dpi=1200)
image2.figure.savefig(raw_dir+'speed.png', format='png', dpi=1200)
```
