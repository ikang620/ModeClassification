```python
# leave ramp out figure
raw_dir = "/content/drive/My Drive/ML Data/Offline Mode Classification Result/ramp remove result/"

final_data = pd.DataFrame()
for model in ['LDA', 'NN', 'SVM', 'XGB']:
  result = pd.DataFrame()
  for incline in [1, 2, 3, 4]:
    data_path = raw_dir + model + "_ramp_remove" + str(incline) + ".txt"
    file = pd.read_csv(data_path, ' ',header=None)
    file.iloc[:, 2] = model
    file.iloc[:, 3] = incline
    file = file.iloc[:,2:5]
    result = result.append(file)
  final_data = final_data.append(result)

final_data.iloc[:,2] = (1-final_data.iloc[:,2])*100
header = {2:'Model',3:'Incline Setting', 4:'Classification Error (%)'}
final_data.rename(columns=header, inplace=True)

image = sns.lineplot(x="Incline Setting", y ="Classification Error (%)", hue = "Model", ci=95 ,data = final_data)
sns.set(style="white", context="talk", palette="deep")
sns.set(style="ticks")
image.set(xticks=[1,2,3,4])
sns.despine()
```

```python
# leave stair out figure
raw_dir = "/content/drive/My Drive/ML Data/Offline Mode Classification Result/stair remove result/"

final_data = pd.DataFrame()
for model in ['LDA', 'NN', 'SVM', 'XGB']:
  result = pd.DataFrame()
  for incline in [1, 2, 3, 4]:
    data_path = raw_dir + model + "_stair_remove" + str(incline) + ".txt"
    file = pd.read_csv(data_path, ' ',header=None)
    file.iloc[:, 2] = model
    file.iloc[:, 3] = incline
    file = file.iloc[:,2:5]
    result = result.append(file)
  final_data = final_data.append(result)

final_data.iloc[:,2] = (1-final_data.iloc[:,2])*100
header = {2:'Model',3:'Stair Setting', 4:'Classification Error (%)'}
final_data.rename(columns=header, inplace=True)

image = sns.lineplot(x="Stair Setting", y ="Classification Error (%)", hue = "Model", ci=95 ,data = final_data)
sns.set(style="white", context="talk", palette="deep")
sns.set(style="ticks")
image.set(xticks=[1,2,3,4])
sns.despine()
```
